**RMS Titanic**

  O RMS Titanic foi um navio de passageiros britânico operado pela White Star Line e construído pelos estaleiros da Harland and Wolff em Belfast. Foi a segunda embarcação da Classe Olympic de transatlânticos depois do RMS Olympic e seguido pelo HMHS Britannic. Projetado pelos engenheiros navais Alexander Carlisle e Thomas Andrews, sua construção começou em março de 1909 e ele foi lançado ao mar em maio de 1911. O Titanic foi pensado para ser o navio mais luxuoso e mais seguro de sua época, gerando lendas que era supostamente "inafundável".
  
   A embarcação partiu em sua viagem inaugural de Southampton para Nova Iorque em 10 de abril de 1912, no caminho passando em Cherbourg-Octeville na França e por Queenstown na Irlanda. Ele colidiu com um iceberg às 23h40min do dia 14 de abril e afundou na madrugada do dia seguinte, sendo um dos maiores desastres marítimos em tempos de paz de toda a história.
   
   Seu naufrágio destacou vários pontos fracos de seu projeto, deficiências nos procedimentos de evacuação de emergência e falhas nas regulamentações marítimas da época. Comissões de inquérito foram instauradas nos Estados Unidos e no Reino Unido, levando a mudanças nas leis internacionais de navegação que permanecem em vigor mais de um século depois.

  Os destroços do Titanic foram procurados por décadas até serem encontrados em 1985 por uma equipe liderada por Robert Ballard. Ele se encontra a 3843 m de profundidade e a 650 km ao sudeste de Terra Nova no Canadá. Sua história e naufrágio permaneceram no imaginário popular durante décadas, levando a produção de vários livros e filmes a seu respeito, mais notavelmente o filme Titanic de 1997. Até hoje o Titanic permanece como um dos navios mais famosos da história, com seus destroços atraindo várias expedições de exploração ao longo dos anos.

**Os registros dos passageiros do RMS Titanic**

  Atualmente a falta do preenchimento de informações em um banco de dados é um grande dificultador na hora de se realizar qualquer tipo de análise, esse é um problema que também ocorria no início do século XX:

```python
if tit.isnull().values.any(): #===Verifica se há alguma informação faltando em qualquer célula do dataset
  print ('Quantidade de dados faltantes para cada coluna:')
else:
  print ('Não há dados faltantes')

print('')

tit.isnull().sum()
```

  Felizmente este problema é facilmente contornado com o auxílio da computação.
  
  Observando as informações não disponíveis é possível verificar que não há registro da idade de 177 dos 891 passageiros (aprox 20%). Para estes, será atribuída na coluna "Age" a idade média de todos a bordo. Já para o registro da informação da cabine de cada passageiro a quantidade de dados faltantes é muito maior, 687 dos 890 passageiros (aprox. 75%), como para esta informação não é possível realizar atribuições, a mesma será retirada da lista. Ainda é possível observar que não há registro do porto de embarque de 2 passageiros, para estes, será atribuído na coluna "Embarked" o porto onde mais pessoas embarcaram.

```python
tit.fillna(tit.mean(), inplace = True)
tit.drop(['Cabin'], axis='columns',inplace=True)
tit['Embarked'].fillna(tit['Embarked'].mode()[0], inplace = True)
```

Realizadas essas alterações podemos verificar que a lista de passageiros agora conta apenas com 11 colunas, já que as informações referentes a cabine de cada passageiro foram retiradas. Porém, também é possível observar que não existem mais informações faltantes:

```python
if tit.isnull().values.any():
  print ('Quantidade de dados faltantes para cada coluna:')
else:
  print ('Não há dados faltantes')

print('')

print(tit.isnull().sum())
pd.concat([tit.head(10), tit.tail(8)])
```

Algumas informações não nos dizem muito sobre o perfil dos passageiros a bordo do RMS Titanic, por isso as informações referentes ao número da passagem, o número de irmãos/cônjuge e o número de filhos/pais à bordo de cada passageiro também serão removidas. Além disso, também será realizado a tradução do cabeçalho de cada coluna para português:

```python
tit.drop(['SibSp'], axis='columns',inplace=True)
tit.drop(['Parch'], axis='columns',inplace=True)
tit.drop(['Ticket'], axis='columns',inplace=True)
tit.rename(columns = {'PassengerId': 'IdPassageiro', 'Survived': 'Sobreviveu','Pclass': 'Classe', 'Name': 'Nome'}, inplace = True)
tit.rename(columns = {'Sex': 'Sexo', 'Age': 'Idade','Fare': 'Tarifa', 'Embarked': 'Embarque'}, inplace = True)

pd.concat([tit.head(10), tit.tail(8)])
```

  Feito a tradução dos cabeçalhos das colunas, é possível verificar que as informações na coluna "Sexo" ainda estão em inglês, portanto, também serão traduzidas. Além disso, na coluna "Sobreviveu" são apresentados apenas 0 e 1, sendo 1 para "Sim" e 0 para "Não", esses dados também serão alterados para facilitar a compreensão.

```python
tit['Sexo'].replace('female', 'Feminino', inplace = True)
tit['Sexo'].replace('male', 'Masculino', inplace = True)
tit['Sobreviveu'].replace(0, 'Não', inplace = True)
tit['Sobreviveu'].replace(1, 'Sim', inplace = True)

pd.concat([tit.head(10), tit.tail(8)])
```

**Estatísticas que nos ajudam a entender mais sobre o acidente**

  Agora que temos todas informações dos passageiros e essas informações estão adequadamente formatadas, podemos realizar uma série de análises para entender melhor o perfil dos passageiros à bordo do RMS Titanic, por exemplo, qual o número de passageiros do sexo masculino e do sexo feminino?

```python
tit['Sexo'].value_counts()
```

Conseguimos verificar que a maior parte dos passageiros do RMS Titanic era de homens.

Também é possível verificar a quantidade de passageiros que sobreviveram ao naufrágio:

```python
tit['Sobreviveu'].value_counts()
```

Verificamos que de acordo com essa base de dados 549 passageiros não sobreviveram, porém, esta não é a lista completa de passageiros. De acordo com o wikipedia haviam na verdade cerca de 2200 pessoas à bordo do RMS Titanic, sendo aproximadamente 1300 passageiros. O Naufrágio do RMS Titanic na verdade teve aproximadamente 1500 vítimas fatais.

É possível também combinar essas duas informações e verificar exatamente a distribuição de sobreviventes entre homens e mulheres:

```python
sobreviventes_sexo = tit.groupby(['Sexo']).agg({'Sobreviveu': 'value_counts'})
sobreviventes_sexo
```

É possível observar que a maior parte das mulheres sobreviveu ao naufrágio: 233 mulheres sobreviventes, enquanto 81 mulheres perderam a sua vida. Já para os homens, a situação se inverte, a maioria dos homens a bordo do Titanic não sobreviveu: 468 homens mortos e 109 sobreviventes. Provavelmente isso é observado devido à um código de conduta datado do século XVIII que prioriza o resgate de mulheres e crianças quando há uma situação de risco e recursos limitados, o famoso: "Mulheres e crianças primeiro!"

Isso fica ainda mais evidente se olharmos a proporção de sobreviventes:

```python
proporcao_sexo = sobreviventes_sexo.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum()))
proporcao_sexo
```

  Observa-se que cerca de 75% das mulheres sobreviveram, ou seja, aproximadamente 3 em cada 4 mulheres sobreviveram ao naufrágio. Para os homens esse número é de cerca de 20%, ou seja, aproximadamente 4 em cada 5 homens morreram no naufrágio do MNS Titanic.

  Também podemos entender melhor o perfil de sobreviventes do Titanic avaliando pela idade. É possível identificar qual a idade com o maior número de sobreviventes e de vítimas do acidente:

```python
top_sobreviventes_idade = tit[tit['Sobreviveu']=='Sim']
print('As idades com mais sobreviventes do Titanic foram:')
print(top_sobreviventes_idade['Idade'].value_counts().head(5))

print(' ')

top_vitimas_idade = tit[tit['Sobreviveu']=='Não']
print('As idades com mais vítimas do Titanic foram:')
print(top_vitimas_idade['Idade'].value_counts().head(5))
```

  É possível verificar que tanto para os sobreviventes quanto para as vítimas a idade mais frequente é de 29.7, porém, essa é na verdade a média de idade de todos os passageiros do RMS Titanic e o valor atribuído para os passageiros que não apresentavam a informação de idade na lista de passageiros.

  Se desconsiderarmos os valores que foram atríbuidos, podemos verificar que a idade com mais sobreviventes do RMS Titanic é de 24 anos, 15 pessoas a bordo tinham essa idade e conseguiram sobreviver ao naufrágio, na sequência temos as idades de 36, 35 e 27 anos, todas com 11 sobreviventes cada.

  Para a lista das idades com mais vítimas observamos que 19 das pessoas que morreram no acidente tinham apenas 21 anos, seguido das idades de 28, 25 e 18 anos, que somavam 18, 17 e 17 vítimas, respectivamente.

  Também é possível relacionar informações de idade, sexo e se sobreviveu ou não para identificar, por exemplo, média de idade dos homens sobreviventes, dos homens vitimados e o mesmo para as mulheres:

```python
idade_media_homens_sobreviventes = tit[(tit['Sobreviveu']=='Sim') & (tit['Sexo']=='Masculino')]
print('A média de idade dos homens sobreviventes do Titanic é de:',(round(idade_media_homens_sobreviventes['Idade'].mean(),1)))
print('')

idade_media_homens_vitimas = tit[(tit['Sobreviveu']=='Não') & (tit['Sexo']=='Masculino')]
print('A média de idade dos homens vítimas do Titanic é de:',(round(idade_media_homens_vitimas['Idade'].mean(),1)))
print('')

idade_media_mulheres_sobreviventes = tit[(tit['Sobreviveu']=='Sim') & (tit['Sexo']=='Feminino')]
print('A média de idade das mulheres sobreviventes do Titanic é de:',(round(idade_media_mulheres_sobreviventes['Idade'].mean(),1)))
print('')

idade_media_mulheres_vitimas = tit[(tit['Sobreviveu']=='Não') & (tit['Sexo']=='Feminino')]
print('A média de idade das mulheres vítimas do Titanic é de:',(round(idade_media_mulheres_vitimas['Idade'].mean(),1)))
```

  Podemos verificar que as médidas das idades dos sobreviventes e vítimas separados entre homens e mulheres são bem próximas, porém, esta informação está sofrendo influência da atribuição da média da idade (29,7) aos passageiros os quais não constavam esta informação.

  Outra análise que podemos fazer é a verificação dos sobreviventes homens mais velhos e mais jovens:

```python
idade_min_homem_sobrevivente = tit[(tit['Sobreviveu']=='Sim') & (tit['Sexo']=='Masculino')]
print('A idade do homem mais jovem sobrevivente do Titanic é de:',(idade_min_homem_sobrevivente['Idade'].min()))
print(' ')

idade_min_homem_vitima = tit[(tit['Sobreviveu']=='Não') & (tit['Sexo']=='Masculino')]
print('A idade do homem mais jovem vítima do Titanic é de:',(idade_min_homem_vitima['Idade'].min()))
print(' ')

idade_max_homem_sobrevivente = tit[(tit['Sobreviveu']=='Sim') & (tit['Sexo']=='Masculino')]
print('A idade do homem mais velho sobrevivente do Titanic é de:',(idade_max_homem_sobrevivente['Idade'].max()))
print(' ')

idade_max_homem_vitima = tit[(tit['Sobreviveu']=='Não') & (tit['Sexo']=='Masculino')]
print('A idade do homem mais velho vítima do Titanic é de:',(idade_max_homem_vitima['Idade'].max()))
```

  Olhando as informações obtidas através da última análise é possível verificar que o intervalo de idade dos passageiros homens do RMS Titanic era muito grande, haviam de recém nascidos a idosos embarcados no transatlântico e mesmo priorizando o resgate de mulheres e crianças é possível verificar que houveram crianças vítimas do acidente.

  Vamos realizar a mesma análise, agora com passageiras do sexo feminino que estavam a bordo do Titanic:


```python
idade_min_mulher_sobrevivente = tit[(tit['Sobreviveu']=='Sim') & (tit['Sexo']=='Feminino')]
print('A idade da mulher mais jovem sobrevivente do Titanic é de:',(idade_min_mulher_sobrevivente['Idade'].min()))
print(' ')

idade_min_mulher_vitima = tit[(tit['Sobreviveu']=='Não') & (tit['Sexo']=='Feminino')]
print('A idade da mulher mais jovem vítima do Titanic é de:',(idade_min_mulher_vitima['Idade'].min()))
print(' ')

idade_max_mulher_sobrevivente = tit[(tit['Sobreviveu']=='Sim') & (tit['Sexo']=='Feminino')]
print('A idade da mulher mais velha sobrevivente do Titanic é de:',(idade_max_mulher_sobrevivente['Idade'].max()))
print(' ')

idade_max_mulher_vitima = tit[(tit['Sobreviveu']=='Não') & (tit['Sexo']=='Feminino')]
print('A idade da mulher mais velha vítima do Titanic é de:',(idade_max_mulher_vitima['Idade'].max()))
```

  Com a análise realizada para as passageiras do sexo feminino foi possível verificar que o intervalo de idade entre as mulheres também era bem grande, também indo de recém nascidas a idosas, porém, a mulher mais velha a bordo do Titanic era 17 anos mais jovem que o homem mais velho.

  Podemos entrar em maiores detalhes para verificar o perfil dos sobreviventes prioritários na hora da evacução do RMS Titanic. Eram classificados como prioritários mulheres e crianças até 15 anos, olhando somente para essa seleção podemos determinar a proporção de prioritários na evacuação do RMS Titanic:


```python
prioridade_mulheres = tit[(tit['Sexo']=='Feminino')]
prioridade_criancas = tit[(tit['Idade']<15)]

prioritarios = pd.concat([prioridade_mulheres, prioridade_criancas])
prioritarios.drop_duplicates(subset="Nome",inplace=True)

proporcao_prioritarios = prioritarios.groupby(['Sexo']).agg({'Sobreviveu': 'value_counts'})
proporcao_prioritarios = proporcao_prioritarios.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum()))
proporcao_prioritarios
```

  Quando analisamos somente a lista dos passageiros classificados como prioritários, podemos observar que aproximadamente 1 a cada 2 meninos de até 15 anos a bordo do RMS Titanic não sobreviveram ao naufrágio do navio, mesmo sendo considerados uma das prioridades na evacuação.

  Podemos comparar esses números com os de sobreviventes que não eram considerados prioritários:    

```python
nao_prioritarios = tit[(tit['Sexo']=='Masculino') & (tit['Idade']>15)]
nao_prioritarios

proporcao_nao_prioritarios = nao_prioritarios.groupby(['Sexo']).agg({'Sobreviveu': 'value_counts'})
proporcao_nao_prioritarios = proporcao_nao_prioritarios.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum()))
proporcao_nao_prioritarios
```

É possível observar que a grande maioria dos que foram classificados como não prioritários (homens com mais de 15 anos) de fato morreram no acidente, porém, ainda assim houveram sobreviventes do grupo não prioritário. Estes deveriam ter cedido seus lugares nos botes salva vidas e priorizado o resgate de 25.8% das mulheres e 46.15% dos meninos de até 15 anos que perderam suas vidas no RMS Titanic. 

  Além das observações baseadas no sexo e na idade dos passageiros, também é possível realizar análises baseadas em outros fatores, como por exemplo, a classe social dos passageiros. Dentro do RMS Titanic, os passageiros foram divididos em 3 classes, sendo a classe 1 a mais exclusiva e luxuosa, a classe 2 sendo uma intermediária e a classe 3 sendo a mais acessível e popular. Vejamos a distribuição dos passageiros por classe a bordo do RMS Titanic:


```python
  tit['Classe'].value_counts()
```

  Observa-se que mais da metade dos passageiros à bordo do RMS Titanic estavam embarcados na classe 3. Podemos verificar a proporção de sobreviventes por classe, dessa forma, é possível avaliar se houve ou não a priorização de alguma classe durante o resgate:

```python
sobreviventes_classe = tit.groupby(['Classe']).agg({'Sobreviveu': 'value_counts'})
sobreviventes_classe
proporcao_classe = sobreviventes_classe.groupby(level=0).apply(lambda x:
                                                 100 * x / float(x.sum()))
proporcao_classe
```

  Observa-se que houve a priorização do resgate dos passageiros baseado na classe social dos mesmos. Aproximadamente 2 a cada 3 passageiros da classe 1 sobreviveram, já para a classe 2 esse valor foi de aproximadamente 1 a cada 2, enquanto isso, na classe 3, apenas 1 a cada 4 passageiros conseguiu sobreviver ao naufrágio do MNS Titanic.
 
  Podemos verificar essa afirmação confrontando o valor médio da tarifa de embarque dos sobreviventes com o valor médio da tarifa de embarque das vítimas:

```python
tarifa_media_vitimas = tit[(tit['Sobreviveu']=='Não')]
print('O valor médio da tarifa de embarque das vítimas do Titanic é de:',(round(tarifa_media_vitimas['Tarifa'].mean(),1)))
print('')

tarifa_media_sobreviventes = tit[(tit['Sobreviveu']=='Sim')]
print('O valor médio da tarifa de embarque dos sobreviventes do Titanic é de:',(round(tarifa_media_sobreviventes['Tarifa'].mean(),2)))
print('')

sns.barplot(x='Sobreviveu',y='Tarifa',data=tit)
plt.show()
```

  Realizando a verificação, observa-se que esses dados corroboram a afirmação de que foi priorizado o resgate de passageiros da classe 1, visto que, em média, o valor pago pela tarifa de embarque dos sobreviventes é maior que o dobro do valor da tarifa de embarque pago pelas vítimas do MNS Titanic.

  Outra comparação possível de se realizar é o valor médio das tarifas de em cada um dos 3 portos de embarque:

```python
tarifa_media_Southampton = tit[(tit['Embarque']=='S')]
print('O valor médio da tarifa de embarque em Southampton é de:',(round(tarifa_media_Southampton['Tarifa'].mean(),2)))
print('')

tarifa_media_Cherbourg = tit[(tit['Embarque']=='C')]
print('O valor médio da tarifa de embarque em Cherbourg é de:',(round(tarifa_media_Cherbourg['Tarifa'].mean(),2)))
print('')

tarifa_media_Queenstown = tit[(tit['Embarque']=='Q')]
print('O valor médio da tarifa de embarque em Queenstown é de:',(round(tarifa_media_Queenstown['Tarifa'].mean(),2)))
print('')

sns.barplot(x='Embarque',y='Tarifa',data=tit)
plt.show()
```

  Existiu uma grande diferença nos valores médios das tarifas de embarque entre os portos. O embarque em Queenstown custou, em média, aproximadamente metade do valor pago pelo embarque em Southampton, que por sua vez, também custou aproximadamente metade do valor médio do embarque em Cherbourg.

  Também é possível avaliar quantas pessoas embarcaram em cada um desses portos:

```python
embarques_Southampton = tit[(tit['Embarque']=='S')]
print((tarifa_media_Southampton['Nome'].count()),' pessoas embarcaram em Southampton.')
print('')

embarques_Cherbourg = tit[(tit['Embarque']=='C')]
print(round(tarifa_media_Cherbourg['Nome'].count()),' pessoas embarcaram em Cherbourg.')
print('')

embarques_Queenstown = tit[(tit['Embarque']=='Q')]
print(round(tarifa_media_Queenstown['Nome'].count()),' pessoas embarcaram em Queenstown.')
print('')

sns.countplot(x='Embarque', data=tit)
plt.show()
```
